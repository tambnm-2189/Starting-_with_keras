```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame
```

```python
from keras import layers
from keras import models
from keras import optimizers
from keras.datasets import mnist
from keras.utils import to_categorical
```

# Getting and preprocessing data

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
#take a glance of data.
data = [train_images, test_images, train_labels, test_labels]
# data shape, dtype
data_shape = DataFrame(index= ['train_images', 'test_images', 'train_labels', 'test_labels'],
                       data= {'shape': [i.shape for i in data],
                              'dtype': [i.dtype for i in data]})
```

```python
data_shape
```

```python
# reshape data to desire input's shape (batch, input dim), output's shape (batch, # class)
train_imgs = train_images.reshape(60000, -1).astype('float')
test_imgs = test_images.reshape(10000, -1).astype('float')
train_labs = to_categorical(train_labels).astype('float')
test_labs = to_categorical(test_labels).astype('float')
```

```python
print('onehot train_labels \n {}'.format(train_labs[:6]))
print('\n train_imgs shape {}'.format(train_imgs.shape))
```

```python
# the range of input array's values are from 0-255. thisn't good for optimizing
# normlalize input data to get a range from [0, 1] or with mean 0, var 1
# input normalization truly makes a big improvement in this fairly simple problem
# as far as i know, it helps alliviate the ill-conditioning of hessian matrix, result in better optimization

# i try normalize to get mean 0 and standard var first, 
#but the result is little worse than normalize by devide to 255 
#(could be because of difference in random initialization)

epsilon = 1e-7
images = np.concatenate((train_imgs, test_imgs), axis = 0)
normalized_images = (images - np.mean(images, axis = 0))/(np.std(images, axis = 0) + epsilon)
#train_imgs = normalized_images[:train_images.shape[0]]
#test_imgs = normalized_images[train_images.shape[0]:]

# so i stick with the simple one: deviding by 255
train_imgs = train_imgs/255
test_imgs = test_imgs/255
```

```python
#plot the difference between an original image and a normalized one
fig1, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(train_images[0])
ax1.set_title('origial image')

ax2.imshow(normalized_images[0].reshape(28, 28))
ax2.set_title('normalized image (mean 0, std 1)')
plt.plot()
```

```python
## Emperical distribution of labels (count #examples of each class)
# a quick check for a skew emperical distribution in labels

def count_labels(labels):
    "return a dictionary of label counts"
    label_counts = dict()
    for i in labels:
        if i not in label_counts.keys():
            label_counts[i] = 1
        else: label_counts[i] += 1
    return label_counts
    
print(sorted(count_labels(train_labels).items()))
print(sorted(count_labels(test_labels).items()))
```

# Build the architecture

```python
###1. specify input shape  : [None, 784], output shape: [None, 10]

###2 build the architect (1 hidden layer with of 32 units)

## using Sequential class ( only use for linear stack of layers)
# create an model instance
model1 = models.Sequential()

# add layers  
model1.add(layers.Dense(units= 128, activation= 'relu', input_shape = (784,)))
model1.add(layers.Dense(units= 10, activation= 'softmax'))
```

```python
## using functional API (allow to build arbitrary architectures)

# create an input instance:
input_tensor = layers.Input(shape= (784,))

# maipulate data tensors
hidden_tensor = layers.Dense(units= 32, activation= 'relu')(input_tensor)
output_tensor = layers.Dense(units= 10, activation = 'softmax')(hidden_tensor)

# create a moodel object instance with inputs is input-tensor and outputs is ouput_tensor
model2 = models.Model(inputs = input_tensor, outputs = output_tensor)
```

# Complication

```python
# this step is same for either model created from Sequential class or Functional API
model1.compile(optimizer= optimizers.RMSprop(learning_rate= 0.008),
              loss= 'categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
model1.summary()
```

# Train

```python
history1 = model1.fit(x = train_imgs, y = train_labs,
           batch_size= 256, epochs= 10, validation_split= 0.1)
```

```python
def train_val_result(_history):
    ## plot some result
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5))

    # plot train and val loss
    ax1.plot(_history.history['loss'], label = 'Train_loss')
    ax1.plot(_history.history['val_loss'], label = 'Val_loss')
    ax1.set_title('Train and val loss')
    ax1.set_ylabel('loss')
    ax1.set_label('epochs')
    ax1.legend()

    # plot train and val accuracy
    ax2.plot(_history.history['accuracy'], label = 'Train_acc')
    ax2.plot(_history.history['val_accuracy'], label = 'Val_acc')
    ax2.set_title('Train and val acc')
    ax2.set_ylabel('acc')
    ax2.set_label('epochs')
    ax2.legend()
    plt.show()
```

```python
train_val_result(history1)
```

# how to choose a learning rate for a particular architecture? 

```python
# shuffly pick only a part of training data set (10%)
permutation = np.random.permutation(train_images.shape[0])
x = train_imgs[permutation][:6000]
y = train_labs[permutation][:6000]

# for this simple problem, i only use line search:
lrs = np.linspace(start= 0.0001, stop= 0.2, num = 10)
```

```python

# make a function taking lr as argument, output a model complied with that lr:
def model_trial(lr):
    # just the above code
    model = models.Sequential()

    # add layers  
    model.add(layers.Dense(units= 128, activation= 'relu', input_shape = (784,)))
    model.add(layers.Dense(units= 10, activation= 'softmax'))
    
    model.compile(optimizer= optimizers.RMSprop(learning_rate= lr),
              loss= 'categorical_crossentropy', 
              metrics=['accuracy'])
    return model

def plot_lr(lrs):
    """plot loss function to see the effect of different lrs"""
    models = [model_trial(lr) for lr in lrs]
    histories = [model.fit(x = x, y = y, batch_size =256, verbose = 0, epochs = 10) for model in models]

    
    #plot loss
    for i in range(len(lrs)):
        plt.plot(histories[i].history['loss'][2:], label = 'lr: %0.3f'%(lrs[i]))
        plt.ylabel('loss')
        plt.xlabel('epochs')
        plt.legend()
    plt.show()
```

```python
#actually i plot training loss from epoch 2nd 
#because the range of loss from epoch 1st is very wide, making vusualization harder
# the results (for different random initialization ) shows that lr from 0.01 to 0.04 can be potential

plot_lr(lrs)
```

```python
# keep searching in the interval 0.001 to 0.03
# i'll go with 0.008
# but sth isnt really making sense here. this is a fairly simple model, why must i choose a small lr?
plot_lr(np.linspace(start= 0.001, stop= 0.03, num = 5 ))
```

# Now, i'll try to build much bigger model then try to regularize it

```python
from keras import regularizers
```

```python
### i started with 3 layers, [512, 768, 256] units each, 
#then gradually increase model layers and #units.
model2 = models.Sequential()
model2.add(layers.Dense(units= 512, activation= 'relu', input_shape = (784,), 
                       kernel_regularizer= regularizers.l2(0.001)))
model2.add(layers.Dense(units= 1000, activation= 'relu', kernel_regularizer= regularizers.l2(0.001)))
model2.add(layers.Dense(units= 2000, activation= 'relu', kernel_regularizer= regularizers.l2(0.001)))
model2.add(layers.Dense(units= 500, activation= 'relu', kernel_regularizer= regularizers.l2(0.001)))
model2.add(layers.Dense(units= 10, activation= 'softmax'))

model2.compile(optimizer = optimizers.RMSprop(0.01), 
               loss = 'categorical_crossentropy', metrics = ['accuracy'])
history2 = model2.fit(x = train_imgs, y = train_labs, 
                      batch_size = 256, epochs =5, validation_split = 0.1)
```
